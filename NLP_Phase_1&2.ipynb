```python
stemmer = FarasaStemmer()
sample = """يُشار إلى أن اللغة العربية يتحدثها أكثر من 422 مليون نسمة ويتوزع متحدثوها في المنطقة المعروفة باسم الوطن العربي. \nبالإضافة إلى العديد من المناطق الأخرى المجاورة مثل الأهواز وتركيا وتشاد والسنغال وإريتريا وغيرها. \n وهي اللغة الرابعة من لغات منظمة الأمم المتحدة الرسمية الست."""
print('Before lemmatization: \n' + sample)
stemmed_text = stemmer.stem(sample) 
print('\n------------------------------------------------------------------------------------------------------------------------------------------\n')
print('After lemmatization: \n' + stemmed_text)
```

```python
def lemmatize_stemming(text):
    return stemmer.stem(text)

# Tokenize and lemmatize
def preprocess(text):
    result=[]
    for token in simple_preprocess(text) :
        if token not in stopwords_list and len(token) > 2:
            # TODO: Apply lemmatize_stemming() on the token, then add to the results list
            #result.append(lemmatize_stemming(token))
            result.append(token)
            
    return result
```

```python
document_num = 3895
doc_sample = documents_desc['Description'][0]
print("Original document: ")
words = []
for word in doc_sample.split(' '):
    words.append(word)
print(words)
print("\n\nTokenized and lemmatized document: ")
print(preprocess(doc_sample))
```

```python
for i in range(len(data['Category'])):
      data['Category'][i] = ' '.join(preprocess(data['Category'][i].split(",")[0]))

category  = data.Category.astype('category').cat.codes
```

```python
desc_column= documents_desc['Description']
proecessed_docs=desc_column.map(lambda x: preprocess(x))
```

```python
n_desc = []
for c in range(len(proecessed_docs)): 
    n_desc.append(' '.join(proecessed_docs[c]))
print(len(n_desc))
```

```python
# Build the vectorizer
vect = TfidfVectorizer(ngram_range=(1, 1), max_features=1000, token_pattern=r'\b[^\d\W][^\d\W]+\b').fit(data.Description)
# Create sparse matrix from the vectorizer
m = vect.transform(n_desc)
ds = pd.DataFrame(m.toarray(), columns=vect.get_feature_names_out())
ds['Category'] = category
```

```python
# Split dataset into 70% train, 30% test
x = ds.drop('Category', axis=1).values
y = ds['Category'].values
SEED = 10
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=SEED, shuffle = True)
```

```python
# Instantiate Random Forest Classifier
model = RandomForestClassifier(n_estimators= 50, min_samples_leaf=0.0001, random_state=SEED, n_jobs=-1) 
# Fit model to the training set
model.fit(X_train, y_train)
```

```python
# Predict test set labels
y_pred = model.predict(X_test) 
# Evaluate train and test set accuracy
train_acc = round(metrics.accuracy_score(y_train, model.predict(X_train)) * 100)
test_acc = round(metrics.accuracy_score(y_test, y_pred) * 100)
print('Accuracy score (train set) = %d' %train_acc + '%')
print('Accuracy score (test set) = %d' %test_acc + '%')
```

```python
from collections import defaultdict
dict1=defaultdict(lambda: defaultdict(list))
```

```python
#tf
from collections import Counter
dictionaryForTermDocument=defaultdict(lambda: defaultdict(list)) #Auxilarity Structure
for i in range(len(list1)):
  for x in range(len(list1[i])):
     dictionaryForTermDocument[i]= {}
     dictionaryForTermDocument[i]=Counter(list1[i])
     
```

```python
#tf-idf dict
import math
IDFDict = {}
calculatedIDF = 0
countDoc = 0

for i in range(len(dictionaryForTermDocument)):
  for word in list1[i]:
    for x in range(len(list1)):
      if  word in list1[x]:
        countDoc+=1
    calculatedIDF = ((dictionaryForTermDocument[i][word] / len(dictionaryForTermDocument[i]))*(math.log(((3896)+1)/countDoc+1)))
    IDFDict[word]  = calculatedIDF
    countDoc=0
```
